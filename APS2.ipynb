```python
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
from math import *
import numpy as np
```

## Solução numérica

```python
def temp_dissipador(
    ro,               # Densidade do material []
    c,                # calor especifico []
    k,                # Condutividade t]ermica []
    h,                # Coeficiente de transf por conveccao
    t_amb,            # Temperatura ambiente
    t_base,           # Temperatura da base
    r,                # Raio da base
    l,                # Comprimento do dissipador [m]
    delta_x,          # Delta x
    tempo_simulacao   # Tempo de simulacao
):

    t_sup = 50
    P = 2 * pi * r    # Perimetro
    A = pi * r**2     # Area
    alpha = k / (c * ro)
    delta_t = delta_x**2 / (alpha * (h*P*delta_x**2/(k*A) + 2))
    nx = int(l/delta_x)

    # Matriz inicial
    matrix = np.zeros(nx)
    matrix[:] = t_sup
    matrix[0] = t_base
    matrix[-1] = 25
    
    # Lista de matrizes
    matrix_list = [matrix]
    
    # Main loop

    # Tempo
    for t in range(1, int(tempo_simulacao / delta_t)):
        
        # Matrix construida no instante
        matrix = np.zeros(nx)
        matrix[:] = t_sup
        matrix[0] = t_base
        matrix[-1] = 25
        
        # Colunas
        for i in range(1, nx-1):

            temp_atual = matrix_list[t-1][i]
            temp_atual_x_mais_um = matrix_list[t-1][i + 1]
            temp_atual_x_menos_um = matrix_list[t-1][i - 1]
            
            tempo_espaco = (temp_atual_x_mais_um - 2*temp_atual + temp_atual_x_menos_um) / delta_x**2
            conv = h*P*(temp_atual - t_amb)/(k*A)

            conta = temp_atual + alpha*delta_t*(tempo_espaco - conv)
            
            matrix[i] = conta
        
        matrix_list.append(matrix)
        
    return matrix_list
```

```python
l = 0.3                # Comprimento do dissipador [m]
ro = 2700              # Densidade do material [kg/m^3]
tol = 1e-10            # Tolerância
k = 180                # Condutividade térmica [W/mK]
c = 896                # Calor especifico [J/KgK]
t_amb = 50             # Temperatura amb [C]
t_base = 100           # Temperatura da base [C]
h = 50                 # coeficiente de transferência de calor por convecção [W/m^2K]
r = 0.001              # Raio [m]
delta_x = 1e-3         # Delta X
tempo_simulacao = 100  # Tempo de simulação[s]

# Rodando função
lista = temp_dissipador(
    ro,
    c,
    k,
    h,
    t_amb,
    t_base,
    r,
    l,                
    delta_x,
    tempo_simulacao
)
```

```python
# Plotando resultados numéricos
plt.plot(range(len(lista[-1])), lista[-1])
plt.title("Cálculo numérico (posição x temperatura)")
plt.xlabel("Posição X na aleta (mm)")
plt.ylabel("Temperatura (°C)")
plt.grid()
plt.show()
```

## Solucao analítica

```python
# Support functions

def get_P(R):
    return pi*R*2

def get_A(R):
    return pi*R**2

def get_m(h, P, k, A):
    return sqrt((h*P)/(k*A))

def get_temp_distrib(m, L, x):

    denominador = sinh(m*L)
    
    div1 = (theta_l/theta_b)*sinh(m*x) / denominador
    div2 = sinh(m*(L-x)) / denominador
    
    return div1 + div2

def get_T(temp_distrib):
    return temp_distrib*(t_base - t_amb) + t_amb
```

```python
# Parâmetros
r = 0.001           # Raio aleta (m)
l = 0.3             # Comprimento aleta (m)
h = 50              # Coeficiente de transferência de calor (W/m2/K)
k = 180             # Condutividade térmica do Alumínio-2014 (W/mK)

t_base = 100        # Temperatura da base
t_ext = 25          # Temperatura da extremidade
t_amb = 50          # Temperatura do ambiente

theta_b = t_base - t_amb
theta_l = t_ext - t_amb

P = get_P(r)        # Perímetro
A = get_A(r)        # Área
m = get_m(h, P, k, A)
```

```python
delta_x = 1e-3
matrix_pontos = np.arange(0,l,delta_x)      # Pontos no eixo x da aleta
matrix_temp = np.zeros(len(matrix_pontos))  # Temperatura encontradas

for i in range(len(matrix_pontos)):
    
    temp_distrib = get_temp_distrib(m, l, matrix_pontos[i])
    T = get_T(temp_distrib)
    
    matrix_temp[i] = T
```

```python
# Plotando gráfico analítico
plt.plot(range(len(matrix_temp)), matrix_temp)
plt.title("Cálculo analítico (posição x temperatura)")
plt.xlabel("Posição X na aleta (mm)")
plt.ylabel("Temperatura (°C)")
plt.grid()
plt.plot()
```

```python
# Plotando os dois gráficos juntos para analisar semelhanças e diferenças
plt.figure(figsize=(15,10))
plt.plot(range(len(matrix_temp)), matrix_temp, label="Analítico")
plt.plot(range(len(lista[-1])), lista[-1], label="Numérico")
plt.title("Cálculo analítico (posição x temperatura)")
plt.xlabel("Posição X na aleta (mm)")
plt.ylabel("Temperatura (°C)")
plt.grid()
plt.legend()
plt.plot()
```
